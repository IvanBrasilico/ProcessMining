```python
import pandas
import pm4py

file_path = 'event-log.csv'
```

```python
event_log = pandas.read_csv(file_path)
event_log.head(10)
```

```python
event_log.dtypes
```

```python
event_log.status.unique()
```

```python
num_events = len(event_log)
num_cases = len(event_log.lawsuit_id.unique())
print("Number of events: {}\nNumber of cases: {}".format(num_events, num_cases))
```

```python
fevent_log = pm4py.format_dataframe(event_log, case_id='lawsuit_id', activity_key='movement', 
                                    timestamp_key='date')
start_activities = pm4py.get_start_activities(fevent_log)
end_activities = pm4py.get_end_activities(fevent_log)
print("Start activities: {}\nEnd activities: {}".format(start_activities, end_activities))
```
